```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wine = load_wine()
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data['TARGET'] = wine.target
```

#### Предварительный анализ данных

```python
print('Количество строк = ', data.shape[0])
print('Количество колонок = ', data.shape[1])
```

#### Корреляционный анализ данных 

```python
#Корреляция целевого признака с остальными
data.corr()['TARGET']
```

В рассматриваемом датасете достаточно много колонок, поэтому для более наглядного отображения корреляции нецелевых признаков будем использовать "тепловую карту".

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True, fmt='.3f')
```

Таким образом, мы видим, что целевой признак сильнее всего коррелирует с признаком alсalinity_of_ash.

Между остальными признаками наибольшая зависимость наблюдается между total_phenols и flavanoids с od280/od315_of_diluted_wines, а также между flavanoids и od280/od315_of_diluted_wines
